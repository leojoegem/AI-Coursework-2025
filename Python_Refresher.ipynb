```python
# Question 1



days = int(input("Enter the number of days: "))


seconds = days * 24 * 60 * 60

#result
print(f"{days} days is equal to {seconds} seconds.")
```

```python
# Question 2
import math
#1
radius = float(input("Enter the radius of the sphere: "))

#2
volume = (4/3) * math.pi * (radius ** 3)

# 3
print(f"The volume of a sphere with radius {radius} is {volume:.2f}")
```

```python
# Question 3


# 1
def area_of_square(side):
    return side ** 2

# 2
def perimeter_of_square(side):
    return 4 * side

# 3
side_length = float(input("Enter the side length of the square: "))

# 4
area = area_of_square(side_length)
perimeter = perimeter_of_square(side_length)

# 5
print(f"A square with side length {side_length} has:")
print(f"- Area: {area}")
print(f"- Perimeter: {perimeter}")
```

```python
# Question 4


#1
def check_character_case(ch):
    if ch.isupper():
        return "Uppercase"
    elif ch.islower():
        return "Lowercase"
    else:
        return "Other..please enter a letter only."

#2
char = input("Enter a single character: ")

#3
result = check_character_case(char)
print(f"The character '{char}' is: {result}")
```

```python
# Question 5


#1
x = 0
y = 20

#2
while y >= 6:
    y = y - 4            # subtract 4 from y
    x = x + (2 / y)      # add 2/y to x

#3
print("Final value of x is:", x)
```

```python
# Question 6


#1
values = []

#2
for i in range(5):
    num = float(input(f"Enter value {i+1}: "))
    values.append(num)

#3
average = sum(values) / len(values)

#4
print("The values you entered are:", values)
print(f"The average of the values is: {average:.2f}")
```
